Current dataset has raw numbers (`budget`, `revenue`, `release_date`). Feature engineering is the process of combining or extracting data from these raw columns to create new columns (features) that actually answer the business questions we defined in your `success.md` file.

```python
import pandas as pd
```

```python
df = pd.read_csv('data/cleaned_movies_dataset.csv')

df['release_date'] = pd.to_datetime(df['release_date'])
```

## FEATURE 1: Financial Metrics (Profit & ROI)

```python
df['profit'] = df['revenue'] - df['budget']

df['roi_percentage'] = (df['profit'] / df['budget']) * 100

# A commercially successful film has an ROI > 100%
df['is_commercial_success'] = df['roi_percentage'] > 100

df.head()
```

## FEATURE 2: Time and Seasonality Metrics

```python
# Extracting the Year, Month, and Day of the Week
df['release_year'] = df['release_date'].dt.year
df['release_month'] = df['release_date'].dt.month_name() 
df['release_day'] = df['release_date'].dt.day_name()

df.head(3)
```

## FEATURE 3: Text & Categorical Extraction

The `genres` column usually looks like this: "Action, Science Fiction, Adventure"

For a new filmmaker, the FIRST genre listed is usually the "Primary" genre of the film.

```python
df['primary_genre'] = df['genres'].apply(lambda x: str(x).split(',')[0].strip() if pd.notnull(x) else 'Unknown')

# Count how many genres the movie belongs to (does mixing genres help or hurt?)
df['genre_count'] = df['genres'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0)
```

## FEATURE 4: Creating our "Sweet Spot" Target
Based on the success criteria: ROI > 100% AND vote_average >= 7.0

```python
df['is_sweet_spot'] = (df['roi_percentage'] > 100) & (df['vote_average'] >= 7.0)
```

```python
output_path = 'data/engineered_movies_dataset.csv'
df.to_csv(output_path, index=False)
```
