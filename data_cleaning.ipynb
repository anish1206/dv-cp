```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('data/tmdb_data_small.csv')
print(f"Original dataset shape: {df.shape}")

df.head(3)
```

## Droping unwanted columns

```python
columns_to_drop =['id', 'backdrop_path', 'homepage', 'imdb_id', 'poster_path', 'overview', 'tagline']
df = df.drop(columns=columns_to_drop)

print(f"Dataset shape: {df.shape}")
df.head(3)
```

## Only keeping released movies

```python
df = df[df['status'] == 'Released']

df = df.drop(columns=['status'])

print(f"Dataset shape: {df.shape}")
df.head(3)
```

## Keeping only those moves whos financial data is available

```python
df = df[df['budget'] > 0]
df = df[df['revenue'] > 0]
```

## Only keep movies with enough votes to prove the rating is real.

```python
df = df[df['vote_count'] >= 100]
```

## Convert release_date from a String to a proper Datetime object so we can extract years/months later

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

## Ensure numeric columns are actually numeric

```python
numeric_cols =['budget', 'revenue', 'runtime', 'vote_average', 'vote_count', 'popularity']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

## Handle remaining Missing Values (Nulls) and removing Duplicates is any

```python
df = df.dropna(subset=['release_date', 'genres', 'runtime'])
df = df.drop_duplicates(subset=['title', 'release_date'])
```

## Final cleaned dataset

```python
print(f"Cleaned dataset shape: {df.shape}")

df.head()
```

> TMDB is having financial data on only a small percentage of its movies. Hence going from 60K to ~7K (about 88% reduction)

```python
output_path = 'data/cleaned_movies_dataset.csv'
df.to_csv(output_path, index=False)
```
